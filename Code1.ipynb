```python
#OUTLIERS

#function to find outliers

def find_outliers_IQR(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return outliers
```

```python
numeric_df= df.select_dtypes(include=['number'])
print(numeric_df)
```

```python

outliers = find_outliers_IQR(df['Postal Code'])

print('number of outliers: '+ str(len(outliers)))

print('max outlier value: '+ str(outliers.max()))

print('min outlier value: '+ str(outliers.min()))

outliers
```

```python
#Replace outliers using imputation (mean value)

def impute_outliers_IQR(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   upper = df[~(df>(q3+1.5*IQR))].max()

   lower = df[~(df<(q1-1.5*IQR))].min()

   df = np.where(df > upper,

       df.mean(),

       np.where(

           df < lower,

           df.mean(),

           df

           )

       )

   return df
```

```python
df['Sales'] = impute_outliers_IQR(df['Sales'])
df.describe()['Sales']
```

```python
 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.impute import KNNImputer
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
```

```python
df.drop(['Row ID','Order ID', 'Customer ID','Customer Name', 'Postal Code', 'Product ID','Product Name'], axis='columns', inplace=True )
print(df)
```

```python
#let's check if there is any abnormal values in  the categorical variables:

df_cat=df[['Order Date','Ship Date','Ship Mode', 'Segment','City','State','Country', 'Market','Region','Category', 'Sub-Category','Order Priority']]

for i in df_cat.columns:
    print(df_cat[i].unique())
```

```python
#transform categorical labels to numerical labels

df_cat=df[['Order Date','Ship Date','Ship Mode','Segment','City','State','Country', 'Market','Region','Category', 'Sub-Category','Order Priority']]

encoders={}

for col_name in df_cat.columns:
    series = df_cat[col_name]
    label_encoder = LabelEncoder()
    df_cat[col_name] = pd.Series(label_encoder.fit_transform(series[series.notnull()]),
    index=series[series.notnull()].index
    )
    encoders[col_name] = label_encoder
```

```python
# Handling Missing Values using Knn for both categorical and numerical labels 

#Numerical values

df_num=df[['Sales','Quantity','Discount','Profit','Shipping Cost']]
imputer = KNNImputer(n_neighbors=5)
df_num.loc[:]=imputer.fit_transform(df_num)

#categorical values

imputer = KNNImputer(n_neighbors=1)
df_cat.loc[:]=imputer.fit_transform(df_cat)

#concatenate the data

df=pd.concat([df_cat,df_num], axis=1)
```

```python
#Load Data
pca = PCA(2)

#Transform the data

d_f= pca.fit_transform(df)
 
d_f.shape
```

```python
# Decode categorical data because k-prototype works with raw categorical data

for i in df_cat.columns:
     df[i]=df[i].astype(int)
for col_name in df_cat.columns:
    df[col_name] = encoders[col_name].inverse_transform(df[col_name])
```

```python
df.info()
```

```python
# Get the position of categorical columns

catColumnsPos = [df.columns.get_loc(col) for col in list(df.select_dtypes('object').columns)]

print('Categorical columns           : {}'.format(list(df.select_dtypes('object').columns)))
print('Categorical columns position  : {}'.format(catColumnsPos))
```

```python
# Convert dataframe to matrix

dfMatrix = df.to_numpy()
```

```python
#Elbow method to detect number of K

from kmodes.kprototypes import KPrototypes

cost = []
for cluster in range(1, 10):
    try:
        kprototype = KPrototypes(n_jobs = -1, n_clusters = cluster, init = 'Huang', random_state = 0)
        kprototype.fit_predict(dfMatrix, categorical = catColumnsPos)
        cost.append(kprototype.cost_)
        print('Cluster initiation: {}'.format(cluster))
    except:
        break

plt.plot(cost)
plt.xlabel('K')
plt.ylabel('cost')
plt.show
```

```python
# cost (sum distance): confirm visual clue of elbow plot
# KneeLocator class will detect elbows if curve is convex; if concave, will detect knees

from kneed import KneeLocator
cost_knee_c3 = KneeLocator(
        x=range(1,10), 
        y=cost, 
        S=0.1, curve="convex", direction="decreasing", online=True)

K_cost_c3 = cost_knee_c3.elbow   
print("elbow at k =", f'{K_cost_c3:.0f} clusters')
```

```python
#let's build the Kprototype model with 3 clusters

kprototype = KPrototypes(n_jobs = -1, n_clusters = 3, init = 'Huang', random_state = 0)
df['clusters']= kprototype.fit_predict(dfMatrix, categorical = catColumnsPos)
```

```python
#predict the labels of clusters.

label = kprototype.fit_predict(dfMatrix, categorical = catColumnsPos)
print(label)
```
